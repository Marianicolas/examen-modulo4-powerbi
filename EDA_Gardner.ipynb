```python
def eliminar_columnas(df):
    columnas_a_eliminar = [
        'HeightofWorkinGardner(cm)*',
        'WidthofWorkinGardner(cm)', 'LengthofText(cm)', 'WidthofText(cm)',
        'ExtraTextLength(cm)', 'ExtraTextWidth(cm)',
        'AreaofWorkinGardner(cm^2)', 'AreaofText(cm^2)', 'ExtraTextArea(cm^2)',
        'TotalAreaText(cm^2)', 'TotalSpace(cm^2)', 'PageArea(cm^2)',
        'SpaceRatioPerPage', 'Book', 'Unnamed: 24', 'Unnamed: 25',
        'Unnamed: 26', 'Unnamed: 27', 'Unnamed: 28', 'Unnamed: 29',
        'Unnamed: 30', 'Unnamed: 31', 'Unnamed: 32', 'Unnamed: 33',
        'Unnamed: 34', 'Unnamed: 35', 'Unnamed: 36', 'Unnamed: 37',
        'Unnamed: 38', 'Unnamed: 39', 'Unnamed: 40', 'Unnamed: 41',
        'Unnamed: 42', 'Unnamed: 43', 'Unnamed: 44', 'Unnamed: 45',
        'Unnamed: 46', 'Unnamed: 47', 'Unnamed: 48', 'Unnamed: 49',
        'Unnamed: 50', 'Unnamed: 51', 'EditionNumber', 'PageNumberofImage', 'ArtistUniqueID '
    ]
    
    # Eliminar columnas que existan en el DataFrame
    df_filtrado = df.drop(columns=[col for col in columnas_a_eliminar if col in df.columns])
    return df_filtrado
```

```python
df_gardner = eliminar_columnas(df_gardner)
```

```python
# Función para aplicar un EDA básico
def eda_basico(df):
    print("========== RESUMEN GENERAL ==========")
    print("Dimensiones:")
    print(F"Este DataFrame tiene {df.shape[0]} filas y {df.shape[1]} columnas." , "\n")
    print()
    print()
    print('-----------------------------------------------------')
    print("Información general:")
    display(df.info())
    print()
    print()
    print('-----------------------------------------------------')
    print("Tipos de datos por columna:")
    print(df.dtypes, "\n")
    print()
    print()
    print('-----------------------------------------------------')
    
    print("\n========== DESCRIBES ==========")
    print("Descripción de columnas numéricas:")
    display(df.describe().T)
    print()
    print()
    print('-----------------------------------------------------')

    print("Descripción de columnas categóricas:")
    try:
        display(df.describe(include=['O']))
    except ValueError:
        print("No se existen columnas categóricas en este DataFrame.")
    print()
    print()
    print('-----------------------------------------------------')

    print("\n========== MUESTRAS ==========")
    print("Primeras filas del DataFrame:")
    display(df.head())
    print()
    print()
    print("Últimas filas del DataFrame:")
    display(df.tail(5))
    print()
    print()
    print("Cinco filas aleatorias del DataFrame:")
    display(df.sample(5))
    print()
    print()
    print('-----------------------------------------------------')
    
    print("\n========== NULOS ==========")
    print("Valores nulos por columna:")

    if df_gardner.isnull().sum().any():
        print("Estas son las columnas en las que hay valores nulos:")
        display(df_gardner.isnull().sum())
    else:
        print("No hay valores nulos en este DataFrame.")

    print("\nDataframe de valores nulos:")
    porcentaje_nulos = (df_gardner.isnull().sum() / df_gardner.shape[0]) * 100
    total_nulos = df_gardner.isnull().sum()

    df_nulos = pd.DataFrame({
        "Total nulos": total_nulos,
        "Porcentaje nulos": porcentaje_nulos,
        "Tipo de dato": df_gardner.dtypes
    })

    df_nulos = df_nulos[df_nulos["Porcentaje nulos"] > 0]
    df_nulos = df_nulos.round(2).reset_index().sort_values(by="Porcentaje nulos", ascending=False)
    df_nulos.columns = ["Columna", "Total nulos", "Porcentaje nulos", "Tipo de dato"]

    display(df_nulos)
    print('-----------------------------------------------------')

    print("\n========== DUPLICADOS ==========")
    dup_count = df.duplicated().sum()
    print(f"Hay {dup_count} filas duplicadas.")

    if dup_count > 0:
        print("\nEstas son las filas duplicadas (sin incluir la primera aparición):")
        display(df[df.duplicated()])

        print("\nEstas son las filas duplicadas incluyendo la primera aparición:")
        display(df[df.duplicated(keep=False)])
    else:
        print("No hay filas duplicadas.")
    
    print('-----------------------------------------------------')
```

```python
eda_basico(df_gardner)
```

```python
#Primer Nulo:
# Localizar el índice del primer nulo en TitleofWork
idx_nulo = df_gardner[df_gardner['TitleofWork'].isnull()].index[0]

# Asignar el nuevo valor
df_gardner.loc[idx_nulo, 'TitleofWork'] = "The Works of Geoffrey Chaucer, now newly imprinted"
```

```python
#Segundo Nulo:
# Localizar el índice del primer nulo en TitleofWork
idx_nulo = df_gardner[df_gardner['TitleofWork'].isnull()].index[0]

# Asignar el nuevo valor
df_gardner.loc[idx_nulo, 'TitleofWork'] = "Letters from Goethe to Lavater from the Years 1774 to 1783"
```

```python
# Para el resto, rellenamos con "Unknown"
for col in ['ArtistNationality', 'ArtistGender', 'ArtistRace', 'ArtistEthnicity']:
    df_gardner[col] = df_gardner[col].fillna('Unknown')
```
